# Example to show the power of genetic algorithms to optmize non-linear equations

In this example, we are going to show the power of genetic algortihms to optmize complicated non-linear functions with several local maximums and local minimums. The chosen function is the [Rastrigin](https://en.wikipedia.org/wiki/Rastrigin_function), which in 2 dimensions is given by:

f(x,y) = A0 + (x)^2 - A1cos(2pi x) + y^2 - A2cos(2pi y);

in the particular case of A0=20 and A1=A2=10, f is given by:

f(x,y) = 20 +(x)^2 - 10cos(2pi x) + y^2 - 10cos(2pi y);


As x^2 and y^2 are always positive, it is clear that the global minimums occurs at (x,y)=(0,0).

**Step 1: Lets take a look at the problem at hand.** 


```
# Import Usefull libraries to plot f(x,y) to get a feeling of the problem

from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import axes3d
import numpy as np
import math as math

# Make data grid for evaluating z=f(x,y)
X = np.arange(-2, 2, 0.01)
Y = np.arange(-2, 2, 0.01)
X, Y = np.meshgrid(X, Y)

Z = 20+ X**2 + Y**2 -10*np.cos(2*math.pi*X) -10*np.cos(2*math.pi*Y)

#Create Figure 

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1, 2, 1, projection='3d')

## Fig name
fig.suptitle('Rastrigin function')

#Surface
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis,
                       linewidth=0, antialiased=False)

plt.xlabel('x')
plt.ylabel('y')
ax.title.set_text('f(x,y)')
plt.axis('off')

## Add color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

#Plot range
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])


## Contour
ax = fig.add_subplot(1, 2, 2)
ax.contourf(X, Y, Z)
ax.title.set_text('Contour')
plt.axis('off')
```
